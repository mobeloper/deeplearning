```python

# Function to generate random emails
def generate_email(adjective, noun, first_name, last_name, year, provider):
    formats = [
        # f"{adjective.lower()}@{provider}",
        # f"{noun.lower()}@{provider}",
        # f"{adjective.lower()}.{noun.lower()}@{provider}",
        # f"{adjective.lower()}_{noun.lower()}@{provider}",
        # f"{adjective.lower()}{noun.lower()}@{provider}",
        
        f"{first_name.lower()}@{provider}",
        f"{first_name.lower()}{year}@{provider}",
        
        f"{last_name.lower()}@{provider}",
        f"{last_name.lower()}{year}@{provider}",
        
        f"{first_name.lower()}{last_name[0].lower()}@{provider}",
        f"{first_name[0].lower()}{last_name.lower()}@{provider}",
        f"{first_name.lower()}{last_name[0].lower()}{year}@{provider}",
        f"{first_name[0].lower()}{last_name.lower()}{year}@{provider}",
        
        f"{first_name.lower()}.{last_name.lower()}@{provider}",
        # f"{first_name.lower()}.{last_name.lower()}{year}@{provider}",
        f"{first_name.lower()}_{last_name.lower()}@{provider}",
        # f"{first_name.lower()}_{last_name.lower()}{year}@{provider}",
        f"{first_name.lower()}{last_name.lower()}{year}@{provider}",
        f"{first_name.lower()}{last_name.lower()}@{provider}"
    ]
    return random.choice(formats)
```
